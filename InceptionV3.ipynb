```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import cv2 as cv
import os
from tensorflow import keras
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from keras.models import Sequential
from sklearn.metrics import precision_score, recall_score, f1_score, roc_curve, auc
from tensorflow.keras import layers, models
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Flatten, Dropout, BatchNormalization
from tensorflow.keras.utils import to_categorical
from keras.models import load_model

#InceptionV3 model
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Flatten, Dense, Dropout
from tensorflow.keras.applications import InceptionV3
import random

train_dir = r'/kaggle/input/animals-detection-images-dataset/train/'
test_dir = r'/kaggle/input/animals-detection-images-dataset/test/'

labels = ['Butterfly',
       'Caterpillar','Chicken', 'Crab',
       'Deer', 'Duck', 'Eagle','Fish',
       'Frog', 'Giraffe','Goose', 'Horse',
       'Jellyfish', 'Ladybug', 'Lion',
       'Lizard','Monkey','Owl', 'Parrot',
       'Penguin',
       'Shark','Snail', 'Snake',
       'Sparrow', 'Spider','Squirrel',
       'Tiger', 'Tortoise','Whale']

label_len = len(labels)
print(label_len)

# Varijable za trening podatke
X = []
Y = []

# Učitavanje podataka
for label in labels:
    folderpath = os.path.join(train_dir, label)
    
    for file in os.listdir(folderpath):
        img_path = os.path.join(folderpath, file)
        img = cv.imread(img_path)
        
        if img is not None:
            img = cv.resize(img, (224, 224))
            X.append(np.array(img))
            Y.append(labels.index(label))

# Konverzija listi u NumPy matrice
X = np.array(X)
Y = np.array(Y)

print("Dimenzije trening podataka:")
print("X shape:", X.shape)
print("Y shape:", Y.shape)

# Varijable za validacijske podatke
X_valid = []
Y_valid = []
X_valid_path = []

for label in labels:
    folderpath = os.path.join(test_dir, label)
    
    for file in os.listdir(folderpath):
        img_path = os.path.join(folderpath, file)
        img = cv.imread(img_path)
        
        if img is not None:
            img = cv.resize(img, (224, 224))
            X_valid.append(np.array(img))
            X_valid_path.append(img_path)
            Y_valid.append(labels.index(label))
            
X_valid = np.array(X_valid)
Y_valid = np.array(Y_valid)

print("\nDimenzije validacijskih podataka:")
print("X_valid shape:", X_valid.shape)
print("Y_valid shape:", Y_valid.shape)

total_images = X.shape[0] + X_valid.shape[0]
print("Broj slika:", total_images)

for i in range(3):
    plt.imshow(X[i])
    plt.title(f"Classe : {Y[i]}")
    plt.show()

    

# Učitavanje prethodno obučenog modela InceptionV3 bez vrha
inception_model = InceptionV3(weights=None, include_top=False, input_shape=(224, 224, 3))

weights_path='/kaggle/input/inception/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5'
inception_model.load_weights(weights_path)
# Zamrzavanje svih slojeva u modelu InceptionV3
for layer in inception_model.layers:
    layer.trainable = False

# Dodavanje potpuno povezanih slojeva na vrh postojećeg modela
x = inception_model.output
x = Flatten()(x)
x = Dense(1024, activation='relu')(x)
x = Dropout(0.5)(x)
predictions = Dense(len(labels), activation='softmax')(x)

# Kombinovanje originalnog modela i novih potpuno povezanih slojeva
inception_model = Model(inputs=inception_model.input, outputs=predictions)

# Kompilacija modela
inception_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Ispisivanje arhitekture modela
inception_model.summary()

# Konverzija oznaka u kategorije one-hot
Y_one_hot = to_categorical(Y)
Y_valid_one_hot = to_categorical(Y_valid)

# Treniranje modela
history_inception = inception_model.fit(X, Y_one_hot, epochs=10, validation_data=(X_valid, Y_valid_one_hot))

# Evaluacija modela
loss_inception, accuracy_inception = inception_model.evaluate(X_valid, Y_valid_one_hot)
print("Test Loss:", loss_inception)
print("Test Accuracy:", accuracy_inception)

# Predikcija na validacijskom skupu
Y_pred_inception = inception_model.predict(X_valid)
Y_pred_classes_inception = np.argmax(Y_pred_inception, axis=1)

# Ispisivanje izveštaja o klasifikaciji
classification_report_inception = classification_report(Y_valid, Y_pred_classes_inception, target_names=labels)
print(classification_report_inception)

# Prikaz grafikona gubitka i tačnosti tokom treninga
plt.plot(history_inception.history['loss'], label='Gubitak na trening skupu')
plt.plot(history_inception.history['val_loss'], label='Gubitak pri validaciji')
plt.title('Trening i Validacijski Gubitak-Istorija')
plt.ylabel('Gubitak')
plt.xlabel('Epohe')
plt.legend()
plt.show()

plt.plot(history_inception.history['accuracy'], label='Tačnost na trening skupu')
plt.plot(history_inception.history['val_accuracy'], label='Tačnost pri validaciji')
plt.title('Trening i Validacijska Tačnost-Istorija')
plt.ylabel('Tačnost')
plt.xlabel('Epohe')
plt.legend()
plt.show()



# Funkcija za prikaz slike i pripadajućih stvarnih i predviđenih klasa
def display_random_images_with_predictions(num_images):
    for _ in range(num_images):
        # Izbor nasumične slike iz validacijskog skupa
        idx = random.randint(0, len(X_valid) - 1)
        image = X_valid[idx]
        true_label = labels[Y_valid[idx]]
        
        # Predikcija klase za izabranu sliku
        prediction = inception_model.predict(np.expand_dims(image, axis=0))
        predicted_label = labels[np.argmax(prediction)]
        
        # Prikaz slike i pripadajućih klasa
        plt.imshow(image)
        plt.title(f"Stvarna klasa: {true_label}, Predviđena klasa: {predicted_label}")
        plt.axis('off')
        plt.show()

# Poziv funkcije za prikaz nasumičnih slika i pripadajućih klasa
display_random_images_with_predictions(3)


# Izračunavanje mikro prosečnih metrika za InceptionV3
micro_precision_inception = precision_score(Y_valid, Y_pred_classes_inception, average='micro')
micro_recall_inception = recall_score(Y_valid, Y_pred_classes_inception, average='micro')
micro_f1_inception = f1_score(Y_valid, Y_pred_classes_inception, average='micro')

print("Mikro prosečne metrike za InceptionV3:")
print("Mikro preciznost:", micro_precision_inception)
print("Mikro odziv:", micro_recall_inception)
print("Mikro F1-score:", micro_f1_inception)

# Izračunavanje makro prosečnih metrika za InceptionV3
macro_precision_inception = precision_score(Y_valid, Y_pred_classes_inception, average='macro')
macro_recall_inception = recall_score(Y_valid, Y_pred_classes_inception, average='macro')
macro_f1_inception = f1_score(Y_valid, Y_pred_classes_inception, average='macro')

print("\nMakro prosečne metrike za InceptionV3:")
print("Makro preciznost:", macro_precision_inception)
print("Makro odziv:", macro_recall_inception)
print("Makro F1-score:", macro_f1_inception)

# Izračunavanje ROC krive i AUC za svaku klasu za InceptionV3
fpr_inception = dict()
tpr_inception = dict()
roc_auc_inception = dict()

for i in range(len(labels)):
    fpr_inception[i], tpr_inception[i], _ = roc_curve(Y_valid_one_hot[:, i], Y_pred_inception[:, i])
    roc_auc_inception[i] = auc(fpr_inception[i], tpr_inception[i])

# Prikazivanje ROC krive za InceptionV3
plt.figure(figsize=(12, 8))
for i in range(len(labels)):
    plt.plot(fpr_inception[i], tpr_inception[i], label=f'{labels[i]} (AUC = {roc_auc_inception[i]:.2f})')

plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlabel('Stopa lažno pozitivnih')
plt.ylabel('Stopa istinito pozitivnih')
plt.title('ROC Kriva za InceptionV3 višeklasnu klasifikaciju')
plt.legend(loc="lower right")
plt.show()
```
